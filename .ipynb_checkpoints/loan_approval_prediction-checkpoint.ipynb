```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn import svm
```

```python
df= pd.read_csv("dataset.csv")
```

```python
df.head()
```

```python
# checking null values
df.isnull().sum()
```

```python
df.shape
```

```python
# checking relations between credit_history and loan_status
sns.countplot(x= df["credit_history"], hue= df["loan_status"], data=df)
```

```python
# getting mathematical data
df.describe()
```

```python
# droping empty values
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
# getting categorical columns and using OneHotEncoder 
cat_cols=["gender", "married", "education", "self_employed", "property_area"]
ohe=OneHotEncoder(sparse_output=False, drop="first")
encoded=ohe.fit_transform(df[cat_cols])
```

```python
# making new DataFrame
df_new= pd.DataFrame(encoded, columns= ohe.get_feature_names_out())
```

```python
# making new dataframe with all values
df2= pd.concat([df.reset_index().drop(cat_cols, axis=1), df_new.reset_index()], axis=1)
```

```python
# droping unnecessary columns
df2.drop(["index", "loan_id"],axis=1, inplace=True)
```

```python
df2
```

```python
# using LabelEncoder in output column
lb=LabelEncoder()
output=lb.fit_transform(df2["loan_status"])
```

```python
df2["loan_status"]=output
```

```python
df2.drop(["property_area_semiurban", "property_area_urban"], axis=1)
```

```python
# checking number of different values in column dependents
df2["dependents"].value_counts()
```

```python
# replacing for better model training 
df2["dependents"].replace("3+", "4", inplace=True)
```

```python
df2.head()
```

```python
df2["dependents"].value_counts()
```

```python
# Labeling Input and output columns
X= df2.drop("loan_status", axis=1)
Y= df2["loan_status"]
```

```python
# making train and test split
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, random_state=2, test_size=0.2, stratify=Y) 
```

```python
clf= svm.SVC(kernel="linear")
```

```python
# accuracy of test data
clf.fit(X_train, Y_train)
y_pred=clf.predict(X_test)
print ("accuracy of test data is : ", accuracy_score(y_pred, Y_test))
```

```python
# accuracy of training data
train_predict= clf.predict(X_train)
print( "accuracy of training data is : ", accuracy_score(Y_train, train_predict))
```
